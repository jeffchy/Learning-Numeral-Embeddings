```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from mpl_toolkits.mplot3d import Axes3D
import pickle
```

```python
numeral_embed_i = pickle.load(open('./p200_i.emb', 'rb'))
numeral_embed_o = pickle.load(open('./p200_o.emb', 'rb'))
```

```python
X_embedded = TSNE(n_components=3, random_state=2).fit_transform(numeral_embed_i).T
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X_embedded[0],X_embedded[1], X_embedded[2], )
a = 0
for x, y, z in zip(X_embedded[0], X_embedded[1], X_embedded[2]):
    ax.text(x, y, z, a)
    a += 1
plt.show()
```
