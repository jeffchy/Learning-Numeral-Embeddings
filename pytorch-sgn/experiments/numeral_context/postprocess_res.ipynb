```python
import json
import numpy as np
import pandas as pd
```

```python
def array(a):
    return np.array(a)

with open('res_TokenNumeral_test.v6.json','r') as f:
    s = f.read()
    s = eval(s)
    print(s)
```

```python
res = {}
res_list = []
score = 5 # 1,2,3,4,5=r
# v = s
for score in [2]:
    
    for k, v in s.items():
        print(k)
#         if k != '500':
#             continue

        avg_rank = np.array(v['avg_rank'])
        if score == 5:
            avg_avg_rank = (np.mean(avg_rank[1:]))
        else:
            avg_avg_rank = (avg_rank[1:][score - 1])

        print(avg_avg_rank)

        mae = np.array(v['mae'])
        if score == 5:
            avg_mae = (np.mean(mae))
        else:
            avg_mae = (mae[score - 1])

        print(avg_mae)

        mdae = np.array(v['mdae'])
        if score == 5:
            avg_mdae = (np.mean(mdae))
        else:
            avg_mdae = mdae[score -1]
        print(avg_mdae)

        mape = np.array(v['mape'])
        if score == 5:
            avg_mape = (np.mean(mape))
        else:
            avg_mape = mape[score -1]

        print(avg_mape)

        mdape = np.array(v['mdape'])
        if score == 5:
            avg_mdape = (np.mean(mdape))
        else:
            avg_mdape = mdape[score -1]
        print(avg_mdape)

        RMSE = np.array(v['RMSE'])
        if score == 5:
            avg_RMSE = (np.mean(RMSE))
        else:
            avg_RMSE = RMSE[score -1]
        print(avg_RMSE)

        lll = [avg_avg_rank, avg_mae, avg_mdae, avg_mape, avg_mdape, avg_RMSE]
        print(lll)
        lll = [i[0] if type(i) == type(np.array([1])) else i for i in lll]
        print(lll)
        lll = [round(i, 2) for i in lll]
        res[k] = lll
        res_list.append(lll)
        
print(np.mean(np.array(res_list), axis=0))
        
```
